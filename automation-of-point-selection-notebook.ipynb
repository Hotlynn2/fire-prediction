# ROST1_NF3

```python
import pandas as pd
import numpy as np
```

```python
file = pd. ExcelFile('boss_nf_sub.xlsx')
rost1_nf3 = pd. read_excel(file, 'rost1_nf3')
rosedale1_nf3 = pd. read_excel(file, 'rosedale_test1')
```

```python
rost1_nf3
```

```python
xrost1nf3cord_list = rost1_nf3['x'].to_list()
# len(xrost1cord_list)
xrost1nf3cord_list
yrost1nf3cord_list = rost1_nf3['y'].to_list()
# len(yrost1cord_list)
yrost1nf3cord_list
```

```python
rosedale1_nf3
```

```python
rosedale1_nf3['target'] = np.where( (rosedale1_nf3['x'].isin(xrost1nf3cord_list) & (rosedale1_nf3['y'].isin(yrost1nf3cord_list) )), 1, 0)
```

```python
rosedale1_nf3.columns
```

```python
rosedale1_nf3 = rosedale1_nf3.drop(['x', 'y', 'xnew', 'ynew', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6',
       'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11',
       'Unnamed: 12', 'xold', 'yold', 'xnewold', 'ynewold'], axis =1)
```

```python
rosedale1_nf3
```

```python
len(rosedale1_nf3.loc[rosedale1_nf3['target']==1])
```
